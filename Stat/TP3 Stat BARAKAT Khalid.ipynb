```
mu = 100.0
sigma = 15.0
data1 = np.linspace(mu - 3*sigma, mu + 3*sigma, 10)
data2 = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
data3 = np.linspace(mu - 3*sigma, mu + 3*sigma, 100000)


plt.plot(data1, norm.pdf(data1, mu, sigma))
plt.show()
print('=================================================')
n = len(data1)
m = data1.mean()
std_err = sem(data1)
h = std_err * t.ppf((1 + 0.95) / 2, n - 1)
print('moyenne echantillon 10 : ', m)
print('ecart-type echantillon 10 : ' , data1.std())
print('erreur-type echantillon 10 : ' ,std_err)
print('interval de confiance echantillon 10 : start: ' ,m - h,' ,end :',m + h)
print('=================================================')


plt.plot(data2, norm.pdf(data2, mu, sigma))
plt.show()
print('=================================================')
n = len(data2)
m = data2.mean()
std_err = sem(data2)
h = std_err * t.ppf((1 + 0.95) / 2, n - 1)
print('moyenne echantillon 100 : ', m)
print('ecart-type echantillon 100 : ', data2.std())
print('erreur-type echantillon 100 : ' , std_err)
print('interval de confiance echantillon 100 : start: ' ,m - h,' ,end :',m + h)
print('=================================================')


plt.plot(data3, norm.pdf(data3, mu, sigma))
plt.show()
print('=================================================')
n = len(data3)
m = data3.mean()
std_100000=data3.std()
std_err = sem(data3)
h = std_err * t.ppf((1 + 0.95) / 2, n - 1)
print('moyenne echantillon 100000 : ' , m)
print('ecart-type echantillon 100000 : ', std_100000)
print('erreur-type echantillon 100000 : ' , std_err)
print('interval de confiance echantillon 100000 : start: ' ,m - h,' ,end :',m + h)
print('=================================================')
```

On remarque que la moyenne est toujours bonne mais que l'cart type n'est pas gale a la valeur thorique, mais plus le nombre de valeurs est lever plus se rapproche de l'cart type voulu.

```
#Implementation de la fonction des coefficients de correlation entre attribut
def correlation(df):
    cor = np.zeros((df.shape[1], df.shape[1]))
    for i in range(df.shape[1]) :
        for j in range(df.shape[1]):
            cor[i][j],_=stats.pearsonr(df.values[:,i],df.values[:,j])
    return cor


print ('Notre Correlation \n',correlation(iris))
print('\nCorrelation predefinie')
print(iris.corr(method ='pearson'))
```

On voie bien que les valeur sont identique a 10^-6 prt.
