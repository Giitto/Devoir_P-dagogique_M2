```
import numpy as np
def E(x):
  return (x-1)*(x-2)*(x-3)*(x-5)

def E_deri(x):
  return 4*x**3 - 33*x**2 + 82*x -61

def DG(mu,x0):
  res = []
  i=0
  
  res.append(x0)
  x1 = res[-1] - mu*E_deri(res[-1])
  i=i+1
  
  while (np.absolute(x1 - res[-1]) > 0.01 and len(res) < 1000):
    i=i+1
    res.append(x1)
    x1 = res[-1] - mu*E_deri(res[-1])
    
  print('Nombre d\'itération :',i)  
  res.append(x1)
  return res

def Ex4(mu,x0):
  print('x0:',x0, ' mu:', mu, ',')
  X = DG(mu,x0)
  m = min(X)
  print('minimum :',m,'\nE(min) :',E(m),'\n\n')
  
```

```
print(DG(0.001,5),'\n')
print(DG(0.01,5),'\n')
print(DG(0.1,5),'\n')
print(DG(0.17,5),'\n')
```

```
#Tres lent
print(DG(1,5),'\n')
```

```
print(DG(0.001,0),'\n')
```

Si on part d'un point x0 (ici x0 = 5) changer la valeur de mu rapproche d'un minima local,

mais si la valeur de mu change trop le descente de gradient va aller vers un autre minima local

Pour x0 = 0 on peut se dire que la descente de gradient est cyclique, hors si on regarde les valeurs dans la table, on voie que les valeurs ne font que monter, le resultat ce rapprocher donc d'un maxima local.

```
import matplotlib.pyplot as plt

x = DG(0.001,5)

lis = []
min = x[0]
ee = []

for i in range(len(x)):
  z = x[i]
  ee.append(E(z))
  if(min <= x[i]):
    lis.append(min)
  else:
    min = x[i]
    lis.append(min)

plt.scatter(ee,lis)
plt.title('Evolution du Minimum en fonction de E(x)')
plt.xlabel('Minimum')
plt.ylabel('E(x)')
plt.show
```

Dans les exemple que j'ai choisi j'ai augmenter le maximum a 10 000 et je divise mu par 10 pour chaque DGi, et on voie bien que le nombre d'itération augmente.

```
def DG1(mu,x0):
  res = []
  i=0
  
  res.append(x0)
  x1 = res[-1] - mu*E_deri(res[-1])
  i=i+1
  
  while (np.absolute(x1 - res[-1]) > 0.001 and len(res) < 10000):
    i=i+1
    res.append(x1)
    x1 = res[-1] - mu*E_deri(res[-1])
    
  print('Nombre d\'itération :',i)  
  res.append(x1)
  return res

def DG2(mu,x0):
  res = []
  i=0
  
  res.append(x0)
  x1 = res[-1] - mu*E_deri(res[-1])
  i=i+1
  
  while (np.absolute(x1 - res[-1]) > 0.0001 and len(res) < 100000):
    i=i+1
    res.append(x1)
    x1 = res[-1] - mu*E_deri(res[-1])
    
  print('Nombre d\'itération :',i)  
  res.append(x1)
  return res

def DG3(mu,x0):
  res = []
  i=0
  
  res.append(x0)
  x1 = res[-1] - mu*E_deri(res[-1])
  i=i+1
  
  while (np.absolute(x1 - res[-1]) > 0.00001 and len(res) < 100000):
    i=i+1
    res.append(x1)
    x1 = res[-1] - mu*E_deri(res[-1])
    
  print('Nombre d\'itération :',i)  
  res.append(x1)
  return res
```
