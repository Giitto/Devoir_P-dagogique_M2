#Header

```
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import *
from sklearn import preprocessing as sk
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
```

#Normalisation de données

```
x = np.matrix([[1,-1,2],[2,0,0],[0,1,-1]], float)
#1 création de la matrice X
print ("x :")
plt.plot(x)
plt.show()
print (x)
#2 Moyenne et variance de X
print ("moyenne :")
print (np.mean(x))
print ("variance :")
print (np.var(x))

print ("\n")
```

```
#3 Normalisation de X
y = sk.scale(x)
plt.plot(y)
plt.show
print (y)
```

**3- on constate, ci-dessu, que la somme des valeurs d'une colonne est égale a  zéro, de plus le représentation visuelle ce ressemble mais ne sont pas assez identique.**

```
#4
print ("moyenne normalise:")
print (np.mean(y))
print ("variance normalise:")
print (np.var(y))
```

**-4 Nous avons normalisé les données pour réduire l'écart entre les valeurs maximal et minimal**

#Normalisation MinMax

```
#création de X2
x2 = np.matrix([[1,-1,2],[2,0,0],[0,1,-1]], float)

print ("x2 :")
print(x2)
plt.plot(x2)
plt.show()
print ("moyenne sur variable:")
print (x2.mean(0))
```

```
print ("Normalisation sur [0 1] :")
y2 = sk.MinMaxScaler((0,1)).fit_transform(x2)
print(y2)
plt.plot(y2)
plt.show()
print ("Moyenne sur variable:")
print (y2.mean(0))
```

**-3 La moyenne sur les variable est presque uniforme (tous comme la matrice X étais presque centré et réduite)**

#Visualisation de données

```
#Chargement des données Iris
iris = datasets.load_iris()
var = iris.feature_names

#Visualisation des donné de Iris
for i in range(len(var)):
  plt.figure(figsize=(20,15))
  for j in range(i+1,len(var)):
    plt.subplot(4,4,j+1)
    plt.scatter(iris.data[:,j],iris.data[:,i], c=iris.target)
    plt.title('Nuage de points avec Matplotlib')
    plt.xlabel(var[i])
    plt.ylabel(var[j])
  plt.show()
```

**-2La meilleur visiualisation est petal lenght comparé a sepal width car on voit une bonne séparation des 3 classes sans corrélation et les valeurs sont continu.**

#Réduction de dimensions et visualisation de données

```
#Création de la donnée en pca et lda
X = iris.data
y = iris.target
target_names = iris.target_names

IrisPca = PCA(n_components=2)
X_r = IrisPca.fit(X).transform(X)

IrisLda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = IrisLda.fit(X, y).transform(X)
```

```
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA de l\'IRIS dataset')

plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA de l\'IRIS dataset')

plt.show()
```

**Pour le PCA nous voyons des données plus proches les une des autres, cela s'explique par le fait que PCA travaille sur l'écart des nuage de données là où LDA travaille sur l'écart des données celon les axes.**
