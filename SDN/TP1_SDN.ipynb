#G√©n√©ration de donn√©es et affichage

Affichage de l'aide pour la fonction make blobs

```
help(make_blobs)
```

Cr√©ation d'une fonction qui utilise figure scatter title xlim ylim xlabel ylabel et show

```
def affichage(x,y,n,t):
  color = ['red','blue','green','yellow']
  plt.figure()
  plt.title(t)
  for i in range(n):
    plt.scatter(x[i,0],x[i,1],c = color[y[i]], s = 20)
  plt.xlim(-15,15)
  plt.ylim(-15,15)
  plt.xlabel('Features')
  plt.ylabel('Value')
  plt.show()
```

G√©n√©ration de 1000 donn√©es de deux variables r√©parties en 4 groupes

```
a,b = make_blobs(n_samples = 1000, n_features = 2, centers = 4)
titre = 'Figure 1000'
affichage(a,b,b.size,titre)
```

G√©n√©ration de 100 donn√©es de deux variables r√©parties en 2 groupes

```
a1,b1 = make_blobs(n_samples = 100, n_features = 2, centers = 2)
titre1 = 'Figure 100'
affichage(a1,b1,b1.size,titre1)
```

G√©n√©ration de 500 donn√©es de deux variables r√©partie en 3 groupes

```
a2,b2 = make_blobs(n_samples = 500, n_features = 2, centers = 3)
titre2 = 'Figure 500'
affichage(a2,b2,b2.size,titre2)
```

Concat√©nation des deux jeux de donn√©es

```
ac,bc = np.vstack((a1,a2)), np.hstack((b1,b2))
titrec = titre1 + ' + '+titre2
affichage(ac,bc,bc.size,titrec)
```
